```python
#Vemos info sobre las columnas
data.columns
```

```python
#Pasamos a español los nombres de las columnas

dicc_columnas = {
    'Row ID': 'ID fila',
    'Order ID':'ID orden',
    'Order Date': 'Fecha de orden',
    'Ship Date': 'Fecha de envío',
    'Ship Mode': 'Modo de envío',
    'Customer ID': 'ID del cliente',
    'Customer Name': 'Nombre del cliente',
    'Segment': 'Segmento',
    'Postal Code': 'Codigo postal',
    'City': 'Ciudad',
    'State': 'Estado',
    'Country': 'País',
    'Region': 'Region',
    'Market': 'Mercado',
    'Product ID': 'ID del producto',
    'Category': 'Categoría',
    'Sub-Category': 'Sub-Categoría',
    'Product Name': 'Nombre del producto',
    'Sales': 'Ventas',
    'Quantity': 'Cantidad',
    'Discount': 'Descuento',
    'Profit': 'Ganancia',
    'Shipping Cost': 'Costo de envío',
    'Order Priority': 'Orden de prioridad'
    
}
```

```python
#Reemplazo el dataset original (columns = dicc_columnas).

data = data.rename(columns = dicc_columnas)
data.head()
```

```python
data.describe()
```

```python
#Convierto las columnas 'Fecha de orden' y 'Fecha de envío' en objetos de fecha y hora
# La función de Pandas tolist() convierte una serie en un tipo de datos de lista. Retorna una lista de valores Series

fecha_orden = data['Fecha de orden'].tolist()
fecha_envio = data['Fecha de envío'].tolist()


print(type(fecha_orden))
```

```python
#Por defecto, al importar el dataset los objetos de fecha y hora se leen como un objeto de cadena, debo
#convertir la cadena de fecha y hora en objeto de fecha y hora de Python. 
#El método to_datetime() hace esto, retornando una serie de objetos fecha y hora

ord_fecha_hora = pd.to_datetime(fecha_orden)
ord_envio = pd.to_datetime(fecha_envio)
```

```python
ord_fecha_hora
```

```python
ord_envio
```

```python
#Después de la conversión anterior (objetos de fecha y hora), los convertimos en objetos de serie
Fecha_orden = pd.Series(ord_fecha_hora, name='Fecha de orden')
Fecha_envio = pd.Series(ord_envio, name='Fecha de envío')
```

```python
#Elimino las columnas que no necesito para responder a las preguntas

data.drop(columns=['ID fila', 'Fecha de orden', 'Fecha de envío', 'ID del cliente', 'Codigo postal'], inplace=True)
```

```python
data.head(2)
```

```python
#El método concat() concatena dos o mas dataframes a lo largo de un eje
data = pd.concat([data, Fecha_orden , Fecha_envio], axis=1)
data.head()
```

```python
data.columns
```

```python
#Con el método reindex cambiamos los índices de las filas y las etiquetas de las columnas.
#Los valores se establecen en NaN si el nuevo índice no es el mismo que el anterior
data = data.reindex(columns = ['ID orden', 'Fecha de orden', 'Fecha de envío', 'Modo de envío', 'Nombre del cliente',
       'Segmento', 'Ciudad', 'Estado', 'País', 'Region', 'Mercado',
       'ID del producto', 'Categoría', 'Sub-Categoría', 'Nombre del producto',
       'Ventas', 'Cantidad', 'Descuento', 'Ganancia',
       'Costo de envío', 'Orden de prioridad'])
```

```python
data.head()
```

```python
#Cambiamos el tipo de dato de la columna Costo de envío con el método astype()
#Pasamos con argumento del método int para especificar el nuevo tipo de dato de la columna
#[:10] --> visualizo sólo las primeras 10 filas
data['Costo de envío'].astype(int)[:10]
```

```python
#Corroboro si todos los valores de las columnas 'Ventas' y 'Ganancias' contienen el símbolo $
#Series.str se usa para acceder a los valores de la serie que son de tipo str y aplicarle métodos. 
#La función Series.str.contains() de Pandas se usa para probar si el patrón o la expresión regular están 
#contenidos dentro de una cadena de una Serie o Índice. 
#Devuelve: una serie o índice de valores booleanos 
```

```python
#La función any() se usa para verificar si algún elemento es Verdadero, potencialmente sobre un eje. Devuelve False si no lo encuentra

if data['Ventas'].str.contains('$').any():
    print("Se encuentra $ en la columna Ventas")
```

```python
if data['Ganancia'].str.contains('$').any:
    print("Se encuentra $ en la columna Ganancia")
```

```python
data.info()
```

```python
#Este atributo devuelve una Serie con el tipo de datos de cada columna.
data.dtypes
```

```python
# Modifico las columnas 'Ventas' y 'Ganancias' que consisten en los símbolos '$'
#El método Series.str.replace() funciona como el método Python .replace() en Series. 
#Devuelve: una serie con valores de texto reemplazados
```

```python
data['Ventas'] = data['Ventas'].str.replace('$', '', regex = True)
data['Ventas'][:10]
```

```python
data['Ventas'] = data['Ventas'].str.replace(',', '', regex = True)
data['Ventas'][:10]
```

```python
data['Ganancia'] = data['Ganancia'].str.replace('$', '', regex = True)
data['Ganancia'][:10]
```

```python
data['Ganancia'] = data['Ganancia'].str.replace(',', '', regex = True)
data['Ganancia'][:10]
```

```python
data['Ganancia'] = data['Ganancia'].str.replace('(', '', regex = True)
data['Ganancia'][:10]
```

```python
data['Ganancia'] = data['Ganancia'].str.replace(')', '', regex = True)
data['Ganancia'][:10]
```

```python
# Después de eliminar la limpieza de las columnas Ventas y Ganancia, las conviérto en columnas numéricas (tipo 'int')
#to_numeric() se usa para convertir el argumento pasado a un tipo numérico.
# El tipo de retorno predeterminado de la función es float64 o int64 según la entrada. 
# Con el parámetro downcast se pueden convertir datos a un tipo particular. Signed equivale al tipo int con signo más pequeño (mín.: np.int8)

data['Ganancia'] = pd.to_numeric(data['Ganancia'], downcast = 'signed')
data['Ganancia'][:10]
```

```python
data['Ventas'] = pd.to_numeric(data['Ventas'], downcast='integer')
data['Ventas'][:10]
```

```python
data['Ganancia'] = data['Ganancia'].astype(int)
data['Ventas'] = data['Ventas'].astype(int)
```
